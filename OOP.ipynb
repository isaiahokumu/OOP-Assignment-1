```python
# Library management system
import csv

class Book:

    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_borrowed = False

    def borrow(self):
        if not self.is_borrowed:
            self.is_borrowed = True
            return True
        return False
    
    def return_book(self):
        if self.is_borrowed:
            self.is_borrowed = False
            return True
        return False
        return False
    
    def __str__(self):
         status = "Borrowed" if self.is_borrowed else "Available"
         return f"Book [{self.isbn}'{self.title}' by {self.author}] - {status}"

class Member:
    def __init__(self, name, member_id):
            self.name = name
            self.member_id = member_id
            self.borrowed_books = []

    def borrow_book(self, book):
        if book.borrow():
            self.borrowed_books.append(book)
            return True
        return False
    
    def return_book(self, book):
        if book in self.borrowed_books:
            book.return_book()
            self.borrowed_books.remove(book)
            return True 
        return False
    
    def __str__(self):
         books = ','.join([book.title for book in self.borrowed_books]) or "None"
         return f"Member [{self.member_id} - {self.name}] - Borrowed Books: {books}"

class Library:
     def __init__ (self):
            self.books = []
            self.members = []
     
     def add_book(self, book):
          self.books.append(book)
          
     def register_member(self, member):
          self.members.append(member)
    
     def find_book(self, isbn):
          for book in self.books:
               if book.isbn == isbn:
                    return book
          return None
     
     def find_member(self, member_id):
          for member in self.members:
               if member.member_id == member_id:
                    return member
          return None
     
     def borrow_book(self, member_id, isbn):
          member = self.find_member(member_id)
          book = self.find_book(isbn)
          if member and book:
              if member.borrow_book(book):
                 print(f"{member.name} borrowed '{book.title}'")
              else:
                  print(f"{book.title} is already borrowed")
          else:
               print("Book or Member not found")
     
     def return_book(self, member_id, isbn):
          member = self.find_member(member_id)
          book = self.find_book(isbn)
          if member and book:
              if member.return_book(book):
                  print(f"{member.name} returned '{book.title}'.")
              else:
                  print(f"{member.name} did not borrow '{book.title}'.")
          else:
               print("Book or Member not found")
     
     def __str__(self):
          book_list = "\n".join(str(book) for book in self.books)
          member_list = "\n".join(str(member) for member in self.members)
          return f"\n--- Library Books---\n{book_list}\n\n---Members---\n{member_list}\n"
     
     def export_books_to_csv(self, filename="books.csv"):
         with open(filename, mode='w', newline='', encoding='utf-8') as file:
             writer = csv.writer(file)
             writer.writerow(["Book No.", "Title", "Author", "Is Borrowed"])
             for book in self.books:
                 writer.writerow([book.isbn, book.title, book.author, book.is_borrowed])

    # Fix export_members_to_csv
     def export_members_to_csv(self, filename='members.csv'):
         with open(filename, mode='w', newline='', encoding='utf-8') as file:
             writer = csv.writer(file)
             writer.writerow(["Member Name", "Member Id", "Borrowed Books(Book No.)"])
             for member in self.members:
                 borrowed_nos = ','.join(book.isbn for book in member.borrowed_books)
                 writer.writerow([member.name, member.member_id, borrowed_nos])

# Defining main function     
def main():
     library = Library()

     #Adding books
     b1 = Book("The River and The Source", "Margaret Ogola", "9789966882059")
     b2 = Book("The Caucasian Chalk Circle", "Bertolt Brecht", "9781559702539")
     b3 = Book("Things Fall Apart", "Chinua Achebe", "9780435272463")

     library.add_book(b1)
     library.add_book(b2)
     library.add_book(b3)

    # Registering members
     mem1 = Member("Timothy", "ZS5243")
     mem2 = Member("Anne", "ZS5246")
     mem3 = Member("Gordon", "ZS5247")

     library.register_member(mem1)
     library.register_member(mem2)
     library.register_member(mem3)

     # initial state
     print("initial state")
     print(library)

     #Borrowed books
     library.borrow_book("ZS5243", "9789966882059")
     library.borrow_book("ZS5246", "9781559702539")
     library.borrow_book("ZS5247", "9780435272463")

     #state after borrowing
     print("\nAfter borrowing:")
     print(library)

     #return a book
     library.return_book("ZS5246", "9781559702539")

     #state after returning the book
     print("\nAfter returning:")
     print(library)

     #Export data to CSV
     library.export_books_to_csv()
     library.export_members_to_csv()
     print("\nLibrary data has been exported to 'books.csv' and 'members.csv'.")

if __name__ == "__main__":
    main()
```
